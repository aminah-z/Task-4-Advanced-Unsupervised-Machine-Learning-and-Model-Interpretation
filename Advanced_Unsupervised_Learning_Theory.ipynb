## Part 1: Unsupervised Machine Learning

### Introduction to Unsupervised Learning

Unsupervised Learning is a type of Machine Learning in which the model learns patterns and relationships from data without using labeled outputs. Instead of predicting a target variable, the algorithm identifies hidden structures or similarities within the dataset.

The two main tasks of unsupervised learning are:
- **Clustering:** Groups similar data points into clusters.
- **Dimensionality Reduction:** Reduces the number of features while preserving important information.

**Example:**
A retail company can use K-Means clustering to group customers based on their purchasing behavior without knowing the customer categories beforehand.

### Clustering vs Dimensionality Reduction

Clustering and Dimensionality Reduction are two important techniques in unsupervised learning, but they serve different purposes.

**Clustering** focuses on grouping similar data points together based on their characteristics. The goal is to discover hidden patterns or segments within the data.

**Dimensionality Reduction** focuses on reducing the number of features while retaining as much important information as possible. This helps simplify complex datasets and improve computational efficiency.

**Example of Clustering:**
A shopping mall groups customers into different segments based on income and spending habits.

**Example of Dimensionality Reduction:**
A dataset containing 100 features can be reduced to 2 principal components using PCA for easier visualization and analysis.

### Distance Metrics

Distance metrics measure how similar or different two data points are. Many clustering algorithms use these metrics to determine which observations should be grouped together.

Some commonly used distance metrics include:

- **Euclidean Distance:** Measures the straight-line distance between two points. It is the most commonly used distance metric in K-Means clustering.
- **Manhattan Distance:** Measures the distance by summing the absolute differences between coordinates. It is useful for grid-like paths.
- **Cosine Similarity:** Measures the angle between two vectors instead of the actual distance. It is commonly used in text analysis and recommendation systems.

**Example:**
When clustering customers based on age and annual income, Euclidean distance can be used to calculate how similar two customers are before assigning them to the same cluster.

### Feature Scaling for Clustering

Feature scaling is an important preprocessing step in clustering because many clustering algorithms calculate the distance between data points. If features have different scales, variables with larger values can dominate the distance calculation and produce inaccurate clusters.

Common feature scaling techniques include:

- **StandardScaler:** Transforms data to have a mean of 0 and a standard deviation of 1.
- **MinMaxScaler:** Scales feature values to a range between 0 and 1.
- **RobustScaler:** Uses the median and interquartile range (IQR), making it more effective for datasets containing outliers.

**Example:**
Consider a dataset with two features: **Age (18–70)** and **Annual Income (5,000–200,000)**. Without scaling, income would have a much greater influence on distance calculations than age. After scaling, both features contribute equally during clustering.

### Curse of Dimensionality

The Curse of Dimensionality refers to the problems that arise when a dataset has a very large number of features. As the number of dimensions increases, data points become more spread out, making it difficult for clustering algorithms to measure similarity accurately. This can lead to increased computational cost and reduced model performance.

To overcome this problem, dimensionality reduction techniques such as PCA, t-SNE, and UMAP are often applied before clustering. These methods reduce the number of features while preserving the most important information.

**Example:**
A dataset with 200 features can be reduced to 20 principal components using PCA before applying K-Means. This reduces computation time and often produces better clustering results.

### Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms the original features into a smaller set of new variables called **principal components**. These components capture most of the variation in the data while reducing redundancy between features.

PCA is commonly used to simplify high-dimensional datasets, improve computational efficiency, and make data easier to visualize. Since it creates new features by combining the original ones, some interpretability may be lost.

**Example:**
A dataset containing 50 features can be reduced to 2 or 3 principal components using PCA. This allows the data to be visualized more easily while retaining most of the important information.

### t-SNE vs UMAP

t-SNE (t-Distributed Stochastic Neighbor Embedding) and UMAP (Uniform Manifold Approximation and Projection) are dimensionality reduction techniques mainly used to visualize high-dimensional data. Both project data into lower dimensions, usually 2D or 3D, making hidden patterns and clusters easier to observe.

t-SNE is effective at preserving local relationships between data points but can be slow for large datasets. UMAP is generally faster, scales better to larger datasets, and preserves both local and some global structure, making it a popular choice for visualization and preprocessing.

**Example:**
When visualizing a dataset with hundreds of features, both t-SNE and UMAP can reveal distinct clusters. However, UMAP often produces similar visualizations in less time, especially for large datasets.

### Cluster Validation Techniques

Cluster validation techniques are used to evaluate the quality of clustering results. Since unsupervised learning does not use labeled data, these techniques measure how well the data points are grouped based on factors such as cluster compactness and separation.

Some commonly used validation metrics include:
- **Silhouette Score:** Measures how similar a data point is to its own cluster compared to other clusters. Higher values indicate better clustering.
- **Davies-Bouldin Index (DBI):** Measures the average similarity between clusters. Lower values indicate better-separated clusters.
- **Calinski-Harabasz Index (CHI):** Evaluates the ratio of between-cluster dispersion to within-cluster dispersion. Higher values indicate better clustering.
- **Inertia (Within-Cluster Sum of Squares):** Measures how close data points are to their assigned cluster centroids. Lower values indicate more compact clusters.

**Example:**
After applying K-Means with different values of **k**, metrics such as Silhouette Score and Inertia can be compared to determine the optimal number of clusters.

## Part 2: Clustering Algorithms

### K-Means Clustering

It groups similar data points into **K** clusters based on their distance from the cluster centroids.
It starts by selecting **K** centroids, assigns each data point to the nearest centroid and then updates the centroids until the clusters no longer change significantly.
Since the number of clusters must be chosen beforehand, methods like,

* **Elbow Method**: determines the optimal number of clusters by plotting the Within-Cluster Sum of Squares (Inertia) for different values of **k**. The point where the decrease in inertia begins to slow down is considered the optimal number of clusters.

* **Silhouette Score** determines the optimal number of clusters by measuring how well data points fit their groups on a scale of -1 to +1. The value of k with the highest score closest to +1 indicates the most distinct and well-separated groupings.

**Example:**
A shopping mall can use K-Means to group customers based on their annual income and spending score, helping the business identify different customer segments for targeted marketing.

### Hierarchical Clustering

Hierarchical Clustering groups data by building a hierarchy of clusters instead of assigning them all at once. Two ways: **Agglomerative**, which starts with individual data points and merges them into larger clusters and **Divisive**, which starts with one large cluster and splits it into smaller ones. The relationships between clusters are often displayed using a **dendrogram**.

**Example:**
A hospital can use Hierarchical Clustering to group patients with similar medical characteristics, helping doctors identify patient groups with similar health conditions.

### DBSCAN

Density-Based Spatial Clustering of Applications with Noise is a clustering algorithm that groups data points based on their density. It identifies dense regions as clusters while labeling isolated points as **noise** or **outliers**. Unlike K-Means, DBSCAN does not require the number of clusters to be specified in advance.
DBSCAN works well for datasets with irregularly shaped clusters but may struggle when cluster densities vary significantly.

**Example:**
In a GPS dataset, DBSCAN can identify areas with high concentrations of taxi pickups while treating isolated pickup locations as outliers.

### Gaussian Mixture Models (GMM)

It assumes the data is generated from a mixture of several Gaussian (normal) distributions. Unlike K-Means, which assigns each data point to a single cluster, GMM calculates the probability that a data point belongs to each cluster, allowing for more flexible clustering.

**Example:**
A bank can use GMM to group customers based on their financial behavior, where some customers may partially belong to multiple spending patterns instead of fitting into a single group.

### Mean Shift

It identifies clusters by repeatedly moving data points toward regions with the highest data density. Unlike K-Means, it does not require the number of clusters to be specified beforehand, as the algorithm automatically determines the clusters based on the data.
Mean Shift works well for discovering clusters of different shapes but can be computationally expensive for large datasets.

**Example:**
In image processing, Mean Shift can group pixels with similar colors to help with image segmentation and object detection.

## Part 3: Dimensionality Reduction

### Principal Component Analysis (PCA)

It is a dimensionality reduction technique
that converts the original features into a smaller set of **principal
components**. These components preserve most of the important information
while reducing the number of features. PCA simplifies high-dimensional data, reduces computational cost, and makes
visualization easier. However, the new components are less interpretable than
the original features.

**Example:**
A dataset with 50 features can be reduced to 2 principal components,
making it easier to visualize and analyze while preserving most of the
data's variation.

### Kernel PCA

Kernel PCA is an extension of PCA that can capture **non-linear**relationships in data. It uses kernel functions t project the data into a higher-dimensional space before reducing its dimensions.Unlike PCA, Kernel PCA can separate complex patterns that cannot becaptured using linear transformations. However, it is more computationally
expensive.

**Example:**
Kernel PCA can be used to reduce the dimensions of image data where the
relationship between features is non-linear, improving visualization and
analysis.

### t-SNE

t-SNE (t-Distributed Stochastic Neighbor Embedding) is a dimensionality
reduction technique mainly used for visualizing high-dimensional data. It
preserves the relationships between nearby data points, making clusters
easier to identify. Although t-SNE produces clear visualizations, it is computationally
expensive and is mainly used for visualization rather than preprocessing.

**Example:**
A handwritten digit dataset with hundreds of features can be reduced to
2 dimensions using t-SNE, making similar digits appear in distinct
clusters.

### UMAP

UMAP (Uniform Manifold Approximation and Projection) is a dimensionality reduction technique that preserves both local and global patterns in the data. It is faster than t-SNE and works well with large datasets. UMAP is commonly used for data visualization and preprocessing because it produces meaningful cluster structures while requiring less computation.

**Example:**
A gene expression dataset with thousands of features can be reduced to
2 dimensions using UMAP, making similar cell types easier to visualize.

## Part 4: Cluster Evaluation Metrics

### Silhouette Score

It measures how well a data point fits within its assigned cluster compared to other clusters. Its value ranges from **-1 to 1**, where a higher score indicates better-defined and well-separated clusters.
- A score close to **1** indicates compact and well-separated clusters.
- A score close to **0** indicates overlapping or poorly separated clusters.

**Example:**
If K-Means produces a Silhouette Score of **0.82**, it indicates that the
clusters are well separated and the clustering quality is good.

### Silhouette Score

It measures how well a data point fits within its assigned
cluster compared to other clusters. Its value ranges from **-1 to 1**,
where a higher score indicates better-defined and well-separated clusters.

- A score close to **1** indicates compact and well-separated clusters.
- A score close to **0** indicates overlapping or poorly separated clusters.

**Example:**
If K-Means produces a Silhouette Score of **0.82**, it indicates that the
clusters are well separated and the clustering quality is good.

### Calinski-Harabasz Index (CHI)

It measures clustering quality by comparing
the separation between clusters with the compactness of data points within
each cluster.

- A **higher CHI** indicates better-defined and well-separated clusters.
- A **lower CHI** suggests weaker clustering performance.

**Example:**
If one clustering model has a CHI of **420** and another has **275**,
the model with **420** is considered to have better clustering quality.

### Inertia (Within-Cluster Sum of Squares)

It measures the total distance between data points and their assigned
cluster centroids. It is commonly used with the **Elbow Method** to help
determine the optimal number of clusters in K-Means.

- **Lower inertia** indicates more compact clusters.
- Inertia should be considered along with other metrics, as a lower value
  does not always mean better clustering.

**Example:**
When testing different values of **K**, the point where inertia begins to
decrease more slowly is selected as the optimal number of clusters using
the Elbow Method.


## Part 5: Model Interpretation

### Feature Importance

It measures how much each feature contributes to a model's predictions. Higher importance features have a greater influence on the model's output, helping identify which variables are most significant. Commonly used to understand machine learning models and support feature selection by highlighting the most relevant variables.

**Example:**
In a customer churn model, **Contract Type** may have a higher feature
importance than **Customer ID**, indicating that it has a greater impact
on predicting customer churn.

### Permutation Importance

It evaluates a feature by randomly shuffling its values and measuring the change in the model's performance. If the performance drops significantly, the feature is considered important. Can be used
with many machine learning models and provides a reliable way to identify
the features that have the greatest impact.

**Example:**
If shuffling the **Age** feature causes a noticeable decrease in model accuracy, it indicates that **Age** is an important feature for prediction.

### SHAP (SHapley Additive exPlanations)

It is a model interpretation technique that explains how each feature contributes to an individual prediction. It assigns a SHAP value to every feature, showing whether it increases or decreases the model's output.

**Example:**
In a loan approval model, SHAP may show that a high income increases the approval probability, while a low credit score decreases it for a particular applicant.

### Partial Dependence Plots (PDP)

It shows how changes in a feature affect a model's predictions while keeping the other features constant. They help visualize the relationship between a feature and the predicted outcome.

**Example:**
In a house price prediction model, a PDP can show how the predicted price changes as the size of the house increases while other features remain unchanged.
